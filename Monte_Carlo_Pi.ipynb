# **What is Monte Carlo?**

Stanislaw Ulam

> Monte Carlo method is a mathematical technique that generates random variables for modelling risk or uncertainty of a certain system. 
>
> The method is used extensively in a wide variety of fields such as physical science, computational biology, statistics, artificial intelligence, and quantitative finance. (The Economic Times)

For the problem we will tackle today we will be drawing random samples from a probability distribution and accepting or rejecting the samples based on some criteria we define. 

# **How to Pull $\pi$ Out of a Square and a Circle:**
---

# **What is The Area of a Square and a Circle?**

![Circle](circle_area.png)

# Area of Square ($A_s$) $= length \cdot width = 2r\cdot 2r = 4r^{2}$ 
# Area of Circle ($A_c$) $= \pi r^{2} $

---

# **Finding a Relation Between The Two Objects Using a Ratio...**



# $$\frac{A_c}{A_s} = \frac{\pi r^2}{2r \cdot 2r} = \frac{\pi r^2}{4r^2} = \frac{\pi}{4}$$

---
# **Solving For $\pi$...**

# $$(4) \cdot \frac{A_c}{A_s} = \frac{\pi}{4} \cdot (4) $$
# $$\pi = 4\frac{A_c}{A_s}$$

---

# **Now Lets Code This in Python**

The idea is to simulate random (x, y) points in a 2-D plane with domain as a square of side 1 unit. Imagine a circle inside the same domain with same diameter and inscribed into the square. We then calculate the ratio of number points that lied inside the circle and total number of generated points.

### Dependencies:

```python
# Dependencies 
import numpy as np 
from numpy import random 
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

### Keeping Track of Our Numbers:

```python
# Number of rocks thrown
number_of_rocks = 30

# Coordinates that are within circle
x_coordinates_in_cirlce = []
y_coordinates_in_circle = []

# Coordinates that are within square 
x_coordinates_in_squarish = []
y_coordinates_in_squarish = []
```

### A While Loop to Simulate The Rocks:

We have to use a uniform distribution for our random numbers as every value is equaly likely to occur in a uniform distribution such as you would think of flipping a coin or rolling a dice. 

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/96/Uniform_Distribution_PDF_SVG.svg/1200px-Uniform_Distribution_PDF_SVG.svg.png" width=500>

```python
count = 1

while count <= number_of_rocks:
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    
    if (x**2 + y**2 <= 1):
        x_coordinates_in_cirlce.append(x)
        y_coordinates_in_circle.append(y)
        
    else:
        x_coordinates_in_squarish.append(x)
        y_coordinates_in_squarish.append(y)
    
    count += 1
    
pi = 4 * len(x_coordinates_in_cirlce)/number_of_rocks

print(pi)
```

### Plot It!

Note: The black dots will be the rocks that land outside the circle while the white dots are will be the rocks that landed inside the circle.

```python
circle = plt.Circle((0, 0), radius = 1, edgecolor = 'yellow',facecolor = 'b', fill = True, lw = 5, ls = '--', alpha = 1)

plt.figure(figsize = (10,10))
ax = plt.gca()

ax1 = plt.plot(x_coordinates_in_cirlce,y_coordinates_in_circle, 'w.')
ax2 = plt.plot(x_coordinates_in_squarish,y_coordinates_in_squarish, 'k.')
ax.add_artist(circle)
ax.add_patch(patches.Rectangle((-1, -1), 2, 2, color = 'green'))
ax.grid()

plt.xlim(-1,1)
plt.ylim(-1,1)
```

### Put it all into a function:

```python
def monte_carlo_pi(number_of_rocks):

    # Coordinates that are within circle
    x_coordinates_in_cirlce = []
    y_coordinates_in_circle = []
    
    # Coordinates that are within square 
    x_coordinates_in_squarish = []
    y_coordinates_in_squarish = []
    
    count = 1

    while count < number_of_rocks:
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        
        if (x**2 + y**2 <= 1):
            x_coordinates_in_cirlce.append(x)
            y_coordinates_in_circle.append(y)
            
        else:
            x_coordinates_in_squarish.append(x)
            y_coordinates_in_squarish.append(y)
        
        count += 1
        
    pi = 4 * len(x_coordinates_in_cirlce)/number_of_rocks
    
    circle = plt.Circle((0, 0), radius = 1, edgecolor = 'yellow',facecolor = 'b', fill = True, lw = 5, ls = '--', alpha = 1)

    plt.figure(figsize = (10,10))
    ax = plt.gca()
    
    ax1 = plt.plot(x_coordinates_in_cirlce,y_coordinates_in_circle, 'w.')
    ax2 = plt.plot(x_coordinates_in_squarish,y_coordinates_in_squarish, 'k.')
    ax.add_artist(circle)
    ax.add_patch(patches.Rectangle((-1, -1), 2, 2, color = 'green'))
    ax.grid()
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    
    return print('The estimated value of Pi is:',pi)    
```

### Pick a number between 100 and 1,000,000:

```python
monte_carlo_pi(1000)
```
